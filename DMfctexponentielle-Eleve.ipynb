```python
from math import *
import matplotlib.pyplot as plt  #importation pour faire des graphiques
import numpy as np    #importation pour la fonction exponentielle graphique

x =np.linspace(0, 10, 100)
y1=66*np.exp(0.34*(x-5))
y2=66
p1=plt.plot(x,y1,color='blue',label='C_f',linewidth=4) #Tracé du dessin représentant la fonction f
p2=plt.plot([0,10],[y2,y2],'r', label="D")  #Tracé de la droite d'équation y=66
plt.xlabel('Années t')
plt.ylabel('Nombre de véhicules électriques')
plt.title("Graphe de la fonction f")
plt.legend()

plt.show()
```
